### Pergunta 1: Tipos de Aprendizagem para o Dataset Iris

O exercício pede para indicar qual tipo de aprendizagem se aplica se quiséssemos prever o resultado de cada coluna isoladamente.  

    sepal length / sepal width / petal length / petal width: Aprendizagem Supervisionada. Justificativa: Se o objetivo explícito é prever o valor numérico contínuo de uma característica em base de outras, precisamos de um histórico para mapear a função preditiva.  

    target: Aprendizagem Supervisionada. Justificativa: É a coluna que identifica a espécie da flor  (o rótulo clássico de destino para classificação) .  

### Pergunta 2: Técnicas Específicas para cada coluna do Iris

Se fôssemos prever cada coluna de forma supervisionada, qual técnica usar?  

    As 4 colunas de medidas (length e width): Regressão. Justificativa: Comprimentos e larguras são valores numéricos contínuos.  

    A coluna target: Classificação. Justificativa: Trata-se de prever uma categoria discreta (as espécies 0, 1 ou 2).  

### Pergunta 3: Cenário dos Estudantes e Flores (Caso dos Perfis/Tamanho)

    Enunciado: Identificar grupos de estudantes por rendimento/proximidade ("alta performance", "baixa autonomia") e clusters de flores por tamanho ("grande/pequeno").  

    Resposta: Agrupamento (Clustering).  

    Justificativa Prática: O objetivo aqui é segmentar as instâncias (linhas). Não estamos filtrando colunas e nem prevendo o rótulo de aprovação, mas sim mapeando perfis de comportamento com base na similaridade matemática das linhas.  

### Pergunta 4: Cenário de Importância de Colunas

    Enunciado: Descobrir se a distância até o campus importa para a aprovação ou se o comprimento da pétala é mais importante para determinar a espécie.  

    Resposta: Seleção de Atributos (Feature Selection).  

    Justificativa Prática: Queremos avaliar o impacto individual de cada coluna original no problema. A seleção de atributos filtra e mantém as colunas intactas, permitindo que a gente explique ao negócio (a administração da PUCPR) se a variável "distância" deve ou não ser descartada do modelo.  

### Pergunta 5: Cenário de Compactação de Colunas

    Enunciado: Unir distância, renda e transporte em uma única coluna de "condição social" ou unir o comprimento e largura da sépala em uma única coluna chamada "sépala".  

    Resposta: Extração de Atributos (Feature Extraction).  

    Justificativa Prática: Ao contrário da seleção, aqui nós transformamos e combinamos os atributos originais para gerar uma nova variável sintética (como os componentes do PCA). Perde-se a leitura direta da coluna original, mas ganha-se uma compactação brutal que preserva a explicabilidade dos dados combinados.
